```python
import pandas as pd
import numpy as np
df=pd.read_excel("MARKLIST.xlsx")
subjects = ['TOC', 'BE', 'DRL', 'AIR', 'GM', 'PROJECT', 'INDUSTRIAL VISIT TRAINING ']
grade_map = {'S': 100, 'A': 90, 'B': 80, 'C': 70, 'D': 60, 'E': 50, 'F': 40}
fail_grades = ['E', 'F']
df.head()
```

```python
for sub in subjects:
    df[sub + '_num'] = df[sub].map(grade_map)
df['Average'] = df[[s + '_num' for s in subjects]].mean(axis=1)
def to_letter(score):
    if score >= 95: return 'S'
    elif score >= 85: return 'A'
    elif score >= 75: return 'B'
    elif score >= 65: return 'C'
    elif score >= 55: return 'D'
    else: return 'E'

df['Avg_Grade'] = df['Average'].apply(to_letter)
df['Has_Fail'] = df[subjects].apply(
    lambda row: any(g in fail_grades for g in row if pd.notna(g)), 
    axis=1
)
df['Result'] = df.apply(
    lambda r: f"FAIL (Avg: {r['Avg_Grade']})" if r['Has_Fail'] else f"PASS (Avg: {r['Avg_Grade']})", 
    axis=1
)
```

```python
result = df[['SNO', 'REG_NO', 'STUDENT_NAME', 'Average', 'Avg_Grade', 'Has_Fail', 'Result']]
print(result.to_string(index=False))
```

```python
df['TOTAL'] = df['Avg_Grade']
df['FAILED?'] = df['Has_Fail'].map({True: 'Y', False: 'N'})

print("✅ TOTAL and FAILED? columns added!")
```

```python
grade_order = {'S': 0, 'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5}
df['Grade_Sort'] = df['TOTAL'].map(grade_order)
df['Fail_Sort'] = df['FAILED?'].map({'N': 0, 'Y': 1})
df = df.sort_values(by=['Fail_Sort', 'Grade_Sort', 'Average'], ascending=[True, True, False])
df = df.reset_index(drop=True)
df['SNO'] = df.index + 1
result = df[['SNO', 'REG_NO', 'STUDENT_NAME', 'TOC', 'BE', 'DRL', 'AIR', 'GM', 
             'PROJECT', 'INDUSTRIAL VISIT TRAINING ', 'TOTAL', 'FAILED?']]
print(result.to_string(index=False))
df.to_excel("MARKLIST_FINAL.xlsx", index=False)
print("\n✅ Saved to MARKLIST_FINAL.xlsx")
```
